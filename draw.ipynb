```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the files
data_1 = np.load('log/test1/tower.test.reward.0.npy')
data_2 = np.load('log/test2/tower.test.reward.0.npy')
data_3 = np.load('log/test3/tower.test.reward.0.npy')
# Plotting the data
plt.figure(figsize=(10, 5))
plt.plot(data_1, label='baselines', marker='o', color='red')
plt.plot(data_2, label='Noise2', marker='o', color='green')
plt.plot(data_3, label='Noise4', marker='o', color='blue')


# Adding labels and title
plt.xlabel('Data Point Index')
plt.ylabel('Value')
plt.title('Line Plot of Two Data Sets')

# Adding legend
plt.legend()

# Showing the plot
plt.grid(True)
plt.show()
```
