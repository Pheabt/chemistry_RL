```python
import numpy as np
import matplotlib.pyplot as plt

# Define the file paths and titles
file_paths = [
    ('log/IID_GRADER_causal_chemistry_chain_1/tower.test.reward.0.npy', 'log/IID_GRADER_causal_chemistry_chain_0/tower.test.reward.0.npy', 'IID_GRADER_causal_chemistry_chain_1'),
    ('log/IID_GRADER_causal_chemistry_collider_1/tower.test.reward.0.npy', 'log/IID_base_GRADER_causal_chemistry_collider_0/tower.test.reward.0.npy', 'IID_GRADER_causal_chemistry_collider_1'),
    ('log/IID_GRADER_causal_chemistry_full_1/tower.test.reward.0.npy', 'log/IID_base_GRADER_causal_chemistry_full_0/tower.test.reward.0.npy', 'IID_GRADER_causal_chemistry_full_1'),
    ('log/IID_GRADER_causal_chemistry_jungle_1/tower.test.reward.0.npy', 'log/IID_base_GRADER_causal_chemistry_jungle_0/tower.test.reward.0.npy', 'IID_GRADER_causal_chemistry_jungle_1'),
    ('log/OOD-S_GRADER_causal_chemistry_chain_1/tower.test.reward.0.npy', 'log/OOD-S_base_GRADER_causal_chemistry_chain_0/tower.test.reward.0.npy', 'OOD-S_GRADER_causal_chemistry_chain_1'),
    ('log/OOD-S_GRADER_causal_chemistry_collider_1/tower.test.reward.0.npy', 'log/OOD-S_base_GRADER_causal_chemistry_collider_0/tower.test.reward.0.npy', 'OOD-S_GRADER_causal_chemistry_collider_1'),
    ('log/OOD-S_GRADER_causal_chemistry_full_1/tower.test.reward.0.npy', 'log/OOD-S_base_GRADER_causal_chemistry_full_0/tower.test.reward.0.npy', 'OOD-S_GRADER_causal_chemistry_full_1'),
    ('log/OOD-S_GRADER_causal_chemistry_jungle_1/tower.test.reward.0.npy', 'log/OOD-S_base_GRADER_causal_chemistry_jungle_0/tower.test.reward.0.npy', 'OOD-S_GRADER_causal_chemistry_jungle_1')
]

# Plot and save the figures for each data group
for i, (file_path_1, file_path_2, title) in enumerate(file_paths):
    # Load the data from the files
    data_1 = np.load(file_path_1)
    data_2 = np.load(file_path_2)

    # Plotting the data
    plt.figure(figsize=(10, 5))
    plt.plot(data_1, label='GRADER', marker='o', color='red')
    plt.plot(data_2, label='OURS', marker='o', color='green')

    # Adding labels and title
    plt.xlabel('Data Point Index')
    plt.ylabel('Value')
    plt.title('Line Plot of ' + title)

    # Adding legend
    plt.legend()

    # Showing the plot
    plt.grid(True)
    plt.savefig(title + ".png")  # Save the plot with the title as the filename
    plt.close()

    print(f"Figure {i+1} saved: {title}.png")
```
