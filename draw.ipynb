```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the files
data_1 = np.load('tower.test.reward.1.npy')
data_2 = np.load('tower.test.reward.2.npy')
data_3 = np.load('tower.test.reward.3.npy')
data_4 = np.load('tower.test.reward.4.npy')
data_5 = np.load('tower.test.reward.5.npy')
data_6 = np.load('tower.test.reward.6.npy')
data_7 = np.load('tower.test.reward.7.npy')


print(data_1)
# Plotting the data
plt.figure(figsize=(10, 5))
plt.plot(data_1, label='Data', marker='o', color='red')
plt.plot(data_2, label='Data', marker='o', color='green')
plt.plot(data_3, label='Data', marker='o', color='blue')
plt.plot(data_4, label='Data', marker='o', color='orange')
plt.plot(data_5, label='Data', marker='o', color='black')
plt.plot(data_6, label='Data', marker='o', color='yellow')
plt.plot(data_7, label='Data', marker='o', color='pink')


# Adding labels and title
plt.xlabel('Data Point Index')
plt.ylabel('Value')
plt.title('Line Plot of Two Data Sets')

# Adding legend
plt.legend()

# Showing the plot
plt.grid(True)
plt.show()
```
